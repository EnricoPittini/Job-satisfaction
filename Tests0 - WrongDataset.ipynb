```python
import pandas as pd
```

```python
df = pd.read_csv('Employee_satisfaction.csv')
```

```python
df.info()
```

```python
df = df[['age', 'Dept', 'location', 'education', 'job_level', 'rating', 'salary', 'satisfied']]
```

```python
dept_dict = {d:i for i,d in enumerate(df['Dept'].unique())}
dept_dict_inv = {i:d for d,i in dept_dict.items()}
```

```python
dept_dict_inv
```

```python
df['location'].unique()
```

```python
location_dict = {l:i for i,l in enumerate(df['location'].unique())}
location_dict_inv = {i:l for l,i in location_dict.items()}
location_dict
```

```python
education_dict = {e:i for i,e in enumerate(df['education'].unique())}
education_dict_inv = {i:e for e,i in education_dict.items()}
education_dict
```

```python
df['Dept'] = df['Dept'].map(dept_dict)
df['location'] = df['location'].map(location_dict)
df['education'] = df['education'].map(education_dict)
```
